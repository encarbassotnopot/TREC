```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("./out_1_avg.json")
df.to_excel("./data.xlsx")
df
```

```python
plt.plot(df)
plt.title("Rendiment dels algoritmes")
plt.xlabel("Nombre d'elements")
plt.ylabel("Temps (s)")
plt.legend(df.columns)
plt.semilogy()
plt.savefig(f"../../img/svg/plot.svg")
plt.savefig(f"../../img/jpg/plot.jpg")
plt.show()
```

```python
for col in df.columns:
	plt.plot(pd.DataFrame(df[col]))
	plt.title(f"Rendiment de {col} sort")
	plt.xlabel("Nombre d'elements")
	plt.ylabel("Temps (s)")
	plt.savefig(f"../../img/svg/{col}.svg")
	plt.savefig(f"../../img/jpg/{col}.jpg")
	plt.show()
```

```python
df = pd.read_json("./out_1_avg.json")

for c in df:
	f = pd.DataFrame(df[c])
	f.columns = ["Temps (s)"]
	f = f.transpose()
	f.to_latex(buf=f'../../latex/tables/times/{c}.tex', na_rep="(-)", longtable=False, float_format="%.4f", bold_rows=True)
```
